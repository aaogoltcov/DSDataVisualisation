```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("owid-covid-data.csv", parse_dates=['date'])
# data[data["location"] == "Russia"]
```

# 1 Bar Charts

### Сравнение стран с максимальным и минимальным количеством случаев заболеваний.
### Вывод: Статистика показывает, что минимальное количество заболеваний приходится на страны среднего востока, максимальное - на Европу. При этом следует отметить, что "открытая" Швеция имеет минимальное количество случаев среди соседей. Количество заболеваний в России находится на достаточно низком уровне

```python
country_barh = data[(data["date"] == "2020-04-24")].sort_values('total_cases_per_million', ascending=True).head(15)
plt.barh(country_barh["location"], country_barh['total_cases_per_million'])
plt.xlabel('Total cases per million')
plt.ylabel('Country')
plt.title("Top 15 most countries with LOW level of coronavirus accedance")
plt.grid()
```

```python
country_barh = data[(data["date"] == "2020-04-24")].sort_values('total_cases_per_million', ascending=False).head(15)
plt.barh(country_barh["location"], country_barh['total_cases_per_million'])
plt.xlabel('Total cases per million')
plt.ylabel('Country')
plt.title("Top 15 most countries with HIGH level of coronavirus accedance")
plt.grid()
```

```python
country_barh = data[((data["date"] == "2020-04-24") & 
                     (data["total_cases_per_million"] > 300) & 
                     (data["total_cases_per_million"] < 600))].sort_values('total_cases_per_million', 
                                                                           ascending=False).head(15)
plt.barh(country_barh["location"], country_barh['total_cases_per_million'])
plt.xlabel('Total cases per million')
plt.ylabel('Country')
plt.title("Top 15 most countries with HIGH level of coronavirus accedance")
plt.text(420, 10, 'Russia', horizontalalignment='left', verticalalignment='center')
plt.grid()
```

# 2 Curves (Plot)

### Сравнение России со страной с похожим уровнем заболеваемости сравнить скорость изменения уровня заболеваемости.
### Вывод: Статистика косвенно указыает на эффективность принимаемых мер со стороны обеих стран. На примере двух стран, превентивные меры в России более эффективны.

```python
dynamic_curves = data[(data["location"] == "Russia") | 
                      (data["location"] == "Bosnia and Herzegovina")].groupby(["date", 
                                                                               "location"]).mean().unstack("location")
dynamic_curves["total_cases_per_million"].plot()
```

# 3 Plot Box

### Сравнение России со страной с похожим уровнем заболеваемости, с помощью Box Plot.
### Вывод: Статистика показывает, что средний уровень заболеваемости в России стремиться к нулю (не считая выборосов), по сравнению с медианой в 5 человек на 1 млн. в Боснии и Герцеговине. Это дополнительно подверждает превентивные меры в России более эффективны.

```python
plot_box = data[(data["location"] == "Russia") | 
                      (data["location"] == "Bosnia and Herzegovina")].groupby(["date", "location"]).mean().unstack("location")
plot_box["total_deaths_per_million"].plot(kind='box')
```

```python
sns.boxplot(data=plot_box["total_deaths_per_million"])
```

# 4 Heatmap

### Heatmap позволяет найти отличия между сравниваемыми объектами.
### Вывод: Наибольшее количество случаев заболевания приходится на Сан Марино, но при этом там достаточно низкий уровень смертности, а в Испании наоборот - количество случае гораздо меньше при максимально высоком уровне смертности. Эта статистика указывает на разные приемы, которые направлены на предотвращение заболеваемости и лечение болезни. Изучая подходы обеих стран потеницально можно найти лучшие практики не только по предтвращению заболеваемости, но и по практике лечения.

```python
comparison_table = pd.pivot_table(data[(data["date"] == "2020-04-24")], index="location", 
                                  values=["total_cases_per_million", 
                                          "total_deaths"], aggfunc=sum).sort_values("total_cases_per_million", 
                                                                                   ascending=False).head(10)
# comparison_table
sns.heatmap(comparison_table, linewidths=0.5)
```

# 5 Pairplot

### Pairplot
### Вывод: Статистика показывает, что количество случаев заболевания не всегда коррелиррует со смертностью в разных странах - таким образом можно определить страны которые лучше всего борятся с болезнью и со смертностью.

```python
deths_plot = data[((data["date"] == "2020-04-24") & 
                     (data["total_cases_per_million"] > 300) & 
                     (data["total_cases_per_million"] < 600))].sort_values('total_cases_per_million', 
                                                                           ascending=False).head(15)
deths_plot = sns.pairplot(deths_plot[['total_cases_per_million', 'total_deaths_per_million']])
deths_plot
```

# 6 Jointplot

### Вывод: В целом, без учета выбросов, с ростом количества заболевших растет и количество смертельных случаев.

```python
join_plot = data[(data["date"] == "2020-04-24") & 
                 (data["total_deaths"] < 1000) & 
                 (data["total_cases"] < 50000)].sort_values('total_cases_per_million', ascending=False)
sns.jointplot(x='total_deaths', y='total_cases', 
              data=join_plot, kind='reg')
```
